### How The Game Works


Below are 6 cards. Each card contains 32 numbers.

The **first number** on the

- first card is 
$2^0 = 1$

- second card is 
$2^1 = 2$

- third card is 
$2^2 = 4$

- fourth card is 
$2^3 = 8$

- fifth card is 
$2^4 = 16$

- sixth card is

**The significance of these first numbers in all the 6 cards:**

Imagine a player thinks of the number 53. The player is shown all the six cards. The player will say that their number exits on cards 1, 3, 5 and 6. So, the computer will add the first numbers on these cards to guess the number chosen by the player.

Therefore, 

$$\text{guess} = 1+4+16+32 = 53$$

Similarly, if the player thinks of the number 28, then they would say that their number exists on the cards 3, 4 and 5. And the computer would add the first numbers on these cards to guess the number chosen by the player.

Therefore, 

$$ \text{guess} = 4+8+16 = 28$$

### Binary Conversion

**How do we know what numbers to fill in each of the 6 cards?** 

I have converted all the numbers between 1 and 63 into its binary notation. A decimal number is expressed as the sum of the product of digits with the powers of 10. For example, 
$29347$ 
is expressed as 

  $$29347 = 2 \times 10^4 + 9 \times 10^3 + 3 \times 10^2 + 4 \times 10^1 + 7 \times 10^0$$

  $$\Rightarrow 29347 = 20000 + 9000 + 300 + 40 + 7$$

Keeping this in mind I have converted a decimal number into its binary notation. For example if we want to convert 29 into its binary notation, then we would have to divide it by 2 until you reach the quotient equal to 0.

- 29, when divided by 2, leaves a quotient 14 and remainder 1.

|Dividend|Divisor|Quotient|Remainder|
|-|-|-|-|
|29|2|14|1|
  
- 14, when divided by 2, leaves a quotient 7 and remainder 0.

|Dividend|Divisor|Quotient|Remainder|
|-|-|-|-| 
|29|2|14|1|
|14|2|7|0|

- 7, when divided by 2, leaves a quotient 3 and remainder 1. 
  
|Dividend|Divisor|Quotient|Remainder|
|-|-|-|-|
|29|2|14|1|
|14|2|7|0|
|7|2|3|1|

- 3, when divided by 2, leaves a quotient 1 and remainder 1.
  
|Dividend|Divisor|Quotient|Remainder|
|-|-|-|-|
|29|2|14|1|
|14|2|7|0|
|7|2|3|1|
|3|2|1|1|

- 1, when divided by 2, leaves a quotient 0 and remainder 1.

|Dividend|Divisor|Quotient|Remainder|
|-|-|-|-|
|29|2|14|1|
|14|2|7|0|
|7|2|3|1|
|3|2|1|1|
|1|2|0|1|

Now, the division process stops because we have reached a quotient equal to 0. The last remainder obtained through the repeated division process is called the **most significant bit (MSB)** and the first remainder obtained is called the **least significant bit (LSB)**.

  So, after arranging the remainders from MSB to LSB, i.e, 
  $11101$
  , I get the binary notation of a decimal number. The number 29 has 5 digits in binary notation, i.e, 
  $1, 1, 1, 0, 1$
  . So to get the decimal number back from its binary notation, the first digit is multiplied with 
  $2^{n - 1} $
  , where 
  $n$ 
  is the number of digits in the binary notation. The second digit is multiplied with 
  $2^{n - 2}$
  , the third digit is multiplied with 
  $2^{n-3}$ 
  and so on 
  $\dots$ 
  The last digit is multiplied with 
  $2^0 = 1$
  . Finally, all these products are added together as shown below.

  $$29 = 1 \times 2^4 + 1 \times 2^3 + 1 \times 2^2 + 0 \times 2^1 + 1 \times 2^0$$

  $$\Rightarrow 29 = 16 + 8 + 4 + 0 + 1$$



### Filling The Cards

Now that I have converted a decimal number into a binary number, I need to put the number into the cards. So, the number 29 will go to cards 5, 4, 3 and 1 because 

  - 16 is the first number on card 5, 
  
  - 8 is the first number on card 4, 
  
  - 4 is the first number on card 3, 
  
  - 0 does not exist on any card, and 
  
  - 1 is the first number on card 1

Similarly, the number 
  $53$ 
  is expressed as 
  $110101$ 
  in binary notation. This is the conversion 
  $110101$ 
  back to its decimal number:
  
  $$53 = 1 \times 2^5 + 1 \times 2^4 + 0 \times 2^3 + 1 \times 2^2 + 0 \times 2^1 + 1 \times 2^0$$
  
  $$\Rightarrow 53 = 32 + 16 + 0 + 4 + 0 + 1$$
  
  So, the number 53 will go to cards 6, 5, 3 and 1 because

  - 32 is the first number on card 6, 

  - 16 is the first number on card 5, 
  
  - 4 is the first number on card 3, and
  
  - 1 is the first number on card 1
